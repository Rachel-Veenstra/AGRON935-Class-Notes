```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
obs_corn_yield = [4500, 5200, 6100, 6600, 7000, 7100, 7000, 6000]
nitrogen_rate = [0,50,100,150,200,250,300,400]
```

```python
plt.scatter(nitrogen_rate, obs_corn_yield)
plt.xlabel('Nitrogen Rates (kg/ha)')
plt.ylabel('Observed Corn Yields (kg/ha)')
plt.show()

par = np.polyfit(nitrogen_rate, obs_corn_yield, 2)
print(par)
```
