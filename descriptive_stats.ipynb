```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
np.random.seed(1)

x = np.linspace(0, 2*np.pi)
y = np.sin(x)
z = y + np.random.randn(y.size)*0.2

plt.plot(x,y)
plt.plot(x,z)
```

```python
print(z.mean())
print(z.std())
print(np.median(z))
```

```python
zpd = pd.Series(z)
print(type(zpd))
```

```python
zpd_smooth = zpd.rolling(5).mean()
zpd_smooth_corrected = zpd.rolling(5, center=True).mean()

plt.plot(x,zpd)
plt.plot(x, zpd_smooth)
plt.plot(x, zpd_smooth_corrected)
plt.plot(x,y)
plt.show()
```
