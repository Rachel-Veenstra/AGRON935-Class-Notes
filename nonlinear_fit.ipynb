```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np

from scipy.optimize import curve_fit

import math
```

```python
np.random.seed(1)

func = lambda x,a,b,c: a*np.exp(-b*x) + c  ## first thing passed must be variable

xdata = np.linspace(0,4,50)

y = func(xdata, 2.5, 1.3, 0.5)

noise = 0.2 * np.random.normal(size=xdata.size)

ydata = y + noise

plt.scatter(xdata, ydata)
```

```python
par_opt, par_cov = curve_fit(func, xdata, ydata)
par_opt
```

```python
plt.scatter(xdata, ydata)

opt_ydata = func(xdata, * par_opt)

plt.plot(xdata, opt_ydata, 'r')
```

```python
SE = np.sqrt(np.mean((ydata-opt_ydata)**2))
SE
```

```python
par_opt_bounded, par_cov_bounded = curve_fit(func, xdata, ydata, bounds=(0, [3., 1., 0.5]))

par_opt_bounded
```

```python
y_opt_bounded = func(xdata, *par_opt_bounded)

plt.scatter(xdata, ydata)

opt_ydata = func(xdata, * par_opt)

plt.plot(xdata, opt_ydata, 'r')
plt.plot(xdata, y_opt_bounded, '--g')
plt.show()
```
