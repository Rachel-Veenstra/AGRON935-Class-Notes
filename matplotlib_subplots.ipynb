```python
# yyplot
plt.figure()
ax1 = plt.subplot()
ax1.plot(df.year, df.Argentina_yield_hg_ha, '--g')

ax2 = ax1.twinx() ## x-axis is shared
ax2.plot(df.year, df.US_yield_hg_ha, '-b')

ax1.set_xlabel('Years')
ax1.set_ylabel('Argentina Yields hg $\mathregular{ha^{-1}}$', color='g')
ax2.set_ylabel('United States Yields hg $\mathregular{ha^{-1}}$', color='b')

plt.show()
```
