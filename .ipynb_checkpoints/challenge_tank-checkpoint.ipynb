```python
import math

vol = 600
capacity = 1500
time = 1
salt = 5
salt_list = [5]
```

```python
while vol <= capacity:
    
    in_salt = (((math.cos(time)) + 1)/5) * 9
    
    vol = vol + 9
    
    out_salt = ((salt+in_salt)/vol) * 6
    
    vol = vol - 6
    
    salt = salt + in_salt - out_salt
    
    time = time + 1
    
    salt_list.append(salt)
    
print(time - 1)
print(salt)
```

```python
import matplotlib.pyplot as plt

plt.plot(salt_list, 'r')
plt.xlabel('Gallons Added')
plt.ylabel('Salt Content (lbs)')
plt.show()
```
