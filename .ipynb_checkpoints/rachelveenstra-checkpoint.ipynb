## Random Walk of Particle

> ##### Inputs: Desired iterations and initial coordinates of particle

> ##### Outputs: Plot of particle's movements over desired number of iterations

#### Author: Rachel Veenstra
#### Created: 04-06-2019

```python
# Importing necessary modules

import random

import matplotlib.pyplot as plt
```

```python
# Creating/assigning variables

iterations = 10000

walk_x = [0]

walk_y = [0]
```

```python
# Creating loop to map particle's walk

for i in range(iterations):
    
    walk_x.append(float(walk_x[i] + random.randint(-1,1))) # particle can move right, left, or neither
    
    walk_y.append(float(walk_y[i] + random.randint(-1,1))) # particle can move up, down, or neither
    
```

```python
# Plotting variables appended to list to show particle's path

plt.plot(walk_x, walk_y, 'k')

plt.ylim(-200, 200)

plt.xlim(-200, 200)

plt.title('Random Walk of Particle - ' + str(iterations) + " Iterations")

plt.show()
```
